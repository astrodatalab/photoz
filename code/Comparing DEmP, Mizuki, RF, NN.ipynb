```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
N = 5
names = ['DEmP', 'Mizuki', 'Random\nforest', 'Neural\nnetwork']
bias = [2.28e-3, 5.8e-3, 1.68e-03, 4.21e-03]
scatter = [2.44e-2, 3.82e-2, 3.17e-02, 3.23e-02]
outlier = [0.09, 0.14, 0.08, 0.06]
loss = [0.245304, 0.254812, 0.128367, 0.097542]
mse = [0.310476, 0.569676, 0.091021, 0.123066]


fig, axes = plt.subplots(3,1, sharex=True, figsize=(5,8))

axes[0].bar(names, bias, color=['indianred', 'coral', 'lightblue', 'yellowgreen', 'coral'], width=0.5,)
axes[0].axhline(y=0.003, ls='--', color='black')
axes[0].set(ylabel='Absolute bias')

axes[1].bar(names, scatter, color=['indianred', 'coral', 'lightblue', 'yellowgreen', 'coral'], width=0.5)
axes[1].axhline(y=0.02, ls='--', color='black')
axes[1].set(ylabel='Scatter')

axes[2].bar(names, outlier, color=['indianred', 'coral', 'lightblue', 'yellowgreen', 'coral'], width=0.5)
axes[2].axhline(y=0.15, ls='--', color='black')
axes[2].set(ylabel='Outlier rate')


fig2, axes2 = plt.subplots(2,1, sharex=True, figsize=(5,6))

axes2[0].bar(names, loss, color=['indianred', 'coral', 'lightblue', 'yellowgreen', 'coral'], width=0.5,)
axes2[0].set(ylabel='Average loss')

axes2[1].bar(names, mse, color=['indianred', 'coral', 'lightblue', 'yellowgreen', 'coral'], width=0.5)
axes2[1].set(ylabel='Mean square error')
```
