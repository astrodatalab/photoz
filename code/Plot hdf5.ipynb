```python
import os
from astropy.io import fits
import tensorflow as tf
import pandas as pd
import numpy as np
import pylab as plt
from sklearn.model_selection import train_test_split
import h5py

import matplotlib

font = {        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
hf = h5py.File('/data/HSC/HSC_v6/five_band_image127x127.hdf5', 'r')
```

```python
hf['image']
```

```python
galaxy_n = [0,1,2,3]
s = hf['image'].shape
plt.figure(figsize=(16,16))
nplot=1
for i in range(len(galaxy_n)):
    ind = galaxy_n[i]
    for j in range(s[1]):
        plt.subplot(len(galaxy_n),s[1],nplot)
        plt.imshow(hf['image'][ind,j,:,:],origin='lower')
        nplot = nplot +1
        plt.xlabel('X Pix')
        plt.ylabel('Y Pix')
plt.tight_layout()
```
