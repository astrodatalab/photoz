```python
import pandas as pd

# Step 1: Read the CSV file using pandas|
csv_file_path = '/data/HSC/shear_catalog/hsc_shear_with_cuts.csv'  # Replace 'input_file.csv' with the path to your CSV file
df = pd.read_csv(csv_file_path)
```

```python
df1 = df.copy()
filt = "HSC-Y"
band = "ytest"

df1.rename(columns={'ira':'ra', 'idec':'dec'}, inplace=True)

# Step 2: Add custom columns with their values
df1['rerun'] = 'pdr2_wide'  # Replace 'rerun' and 'pdr_2' as desired
df1['filter'] = f'{filt}'  # Replace 'filter' and 'HSC-I' as desired
df1['sw'] = '10asec'
df1['sh'] = '10asec'
df1['name'] = df1['object_id'].apply(lambda x: f'{x}_{band}band')

# Step 3: Add an empty descriptor column as the first column
df1.insert(0, '#?', '')

# Step 4: Write the DataFrame to a delimited text file
output_file_path1 = f'/home/jsoriano/shear_code/image_querry/hsc_shape_batch1{band}.txt'  # Replace 'output_file.txt' with the desired output file path
output_file_path2 = f'/home/jsoriano/shear_code/image_querry/hsc_shape_batch2{band}.txt'
output_file_path3 = f'/home/jsoriano/shear_code/image_querry/hsc_shape_batch3{band}.txt'
delimiter = '\t'  # Change to ',' if you want a CSV file

# Shuffle the DataFrame
df1_shuffled = df1.sample(frac=1, random_state=42)
sample1 = df1_shuffled[:2]
sample2 = df1_shuffled[2:4]
sample3 = df1_shuffled[4:6]

#Deleted rerun for test add back later
sample1 = sample1[['#?', 'filter', 'ra', 'dec', 'sw', 'sh', 'name']]
sample2 = sample2[['#?', 'filter', 'ra', 'dec', 'sw', 'sh', 'name']]
sample3 = sample3[['#?', 'filter', 'ra', 'dec', 'sw', 'sh', 'name']]


sample1.to_csv(output_file_path1, sep=delimiter, index=False)
sample2.to_csv(output_file_path2, sep=delimiter, index=False)
sample3.to_csv(output_file_path3, sep=delimiter, index=False)
```
