```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
#import data

photozdata = pd.read_csv('/data/HSC/HSC_IMAGES_FIXED/HSC_photozdata_full_header_trimmed.csv')
```

```python
#filter out bad points

filt =(photozdata['specz_redshift_err'] >0) & (photozdata['specz_redshift_err'] < 1)
photozdata_subset = photozdata[filt]
```

```python
#make training and testing data

y = photozdata_subset['specz_redshift']
X = photozdata_subset[['g_cmodel_mag','r_cmodel_mag','i_cmodel_mag','z_cmodel_mag','y_cmodel_mag']]
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

y_test = np.asarray(y_test)
```

```python
#use the random forest model

randomForest = RandomForestRegressor(n_estimators = 200)
randomForest.fit(X_train, y_train)
```

```python
#generate point estimate

y_test_predict = randomForest.predict(X_test)
```

```python
#generate PDFs

predictions = []
for i in range (len(randomForest.estimators_)):
    predictions.append(
    np.array(
    randomForest.estimators_[i].predict(X_test)))
    
predictions = np.asarray(predictions).T
predictions = np.sort(predictions, axis = 1)
length = len(predictions[0])
```

```python
predictions
```

```python
#find out where all the outlier points are for point estimate and the outlier rate

bias = abs(y_test_predict - y_test)-0.15*(1+y_test)
outlier_point = np.where(bias > 0)[0]

number_outlier_point = len(outlier_point)

outlier_rate_point = number_outlier_point*1.0/len(y_test)

print(number_outlier_point)
print(outlier_rate_point)
```

```python
#calculate the likelihood of true redshift for each galaxy

likelihood = []

for i in range (len(predictions)):
    likelihood.append(
    len(np.where(abs(predictions[i]-y_test[i])<0.15*(1+y_test[i]))[0])/length)

likelihood = np.asarray(likelihood)
```

```python
#plot the likelihood with pdf

plt.figure()
plt.hist(likelihood,bins = 50)
plt.title("likelihood of true redshift with pdf")
plt.show()
```

```python
#outlier rate using pdf

outlier_pdf = np.where(likelihood < 0.6)[0]
number_outlier_pdf = len(outlier_pdf)

outlier_rate_pdf = number_outlier_pdf*1.0/len(likelihood)

print(number_outlier_pdf)
print (outlier_rate_pdf)
```

```python
#find the points that are outliers both with point estimate method and pdf method

intersections = np.intersect1d(outlier_point,outlier_pdf)

for i in range(5):
    n = intersections[i]
    plt.figure()
    plt.hist(predictions[n],bins = 50)
    plt.title("Zspec = %s, Zpredict = %s"%(y_test[n],y_test_predict[n]))
    plt.show()
```

```python
#find points that are outliers with point estimate method but not with pdf methods

outlier_point_notpdf = np.setdiff1d(outlier_point,outlier_pdf)

for i in range(5):
    n = outlier_point_notpdf[i]
    plt.figure()
    plt.hist(predictions[n],bins = 50)
    plt.title("Zspec = %s, Zpredict = %s"%(y_test[n],y_test_predict[n]))
    plt.show()
```

```python
#find points that are outliers with pdf method but not with point estimate methods

outlier_pdf_notpoint = np.setdiff1d(outlier_pdf,outlier_point)

for i in range(5):
    n = outlier_pdf_notpoint[i]
    plt.figure()
    plt.hist(predictions[n],bins = 50)
    plt.title("Zspec = %s, Zpredict = %s"%(y_test[n],y_test_predict[n]))
    plt.show()
```

```python
#points that are not outliers

total = range(len(predictions))
notoutlier_pdf = np.setdiff1d(total,outlier_pdf)
notoutlier = np.setdiff1d(notoutlier_pdf,outlier_point)

for i in range(5):
    n = notoutlier[i]
    plt.figure()
    plt.hist(predictions[n],bins = 50)
    plt.title("Zspec = %s, Zpredict = %s"%(y_test[n],y_test_predict[n]))
    plt.show()
```

```python
#create PIT distribution

PIT  = np.zeros(len(predictions))
for i in range (len(predictions)):          
    PIT[i] = len(np.where(predictions[i]<y_test[i])[0])*1.0/length
```

```python
#plot the histogram of PIT

plt.figure()
plt.hist(PIT, bins = 50)
plt.title("PIT")
plt.show()
```

```python
PIT2  = []
for i in notoutlier_pdf:
    PIT2.append(len(np.where(predictions[i]<y_test[i])[0])*1.0/length)
```

```python
#plot the histogram of PIT

plt.figure()
plt.hist(PIT2, bins = 50)
plt.title("PIT2")
plt.show()
```

```python
#calculate the crps

crps = np.zeros(len(predictions))
for i in range (len(predictions)):
    for j in range (100):
        z = 2.0*j/100
        if z < y[i]:
            crps[i] += ((len(np.where(predictions[i]<z)[0])*1.0/length)**2)*(2.0/100)
        else:
            crps[i] += ((len(np.where(predictions[i]<z)[0])*1.0/length-1)**2)*(2.0/100)     
```

```python
#plot CRPS

CRPS = sum(crps)/len(predictions)
plt.figure()
plt.hist(crps, bins = 50)
plt.title("CRPS = %s"%(CRPS))
plt.show()
```

```python
#plot log crps

crps_log = np.log(crps)
CRPS_log = sum(crps_log)/len(predictions)
plt.figure()
plt.hist(crps_log,bins = 50)
plt.title("log(CPRS)=%s"%(CRPS_log))
plt.show()
```
